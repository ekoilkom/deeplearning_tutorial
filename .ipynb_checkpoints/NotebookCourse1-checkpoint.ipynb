# The Hello World of Deep Learning with Neural Networks

Seperti setiap aplikasi pertama Anda harus mulai dengan sesuatu yang sangat sederhana yang menunjukkan cara kerja kode Anda.

Dalam kasus membuat jaringan saraf, sampel yang saya suka gunakan adalah suatu sistem yang belajar hubungan antara dua angka. Jadi, misalnya, jika Anda menulis kode untuk fungsi seperti dibawah ini:


`` `
float hw_function (float x) {
     float y = (2 * x) - 1;
     return y;
}
`` `

Sehingga jika di berikan

nilai x=-1, maka menghasilkan y=-3

nilai x=0, maka menghasilkan y=-1

nilai x=1, maka menghasilkan y=1

nilai x=2, maka menghasilkan y=3

nilai x=3, maka menghasilkan y=5

nilai x=4, maka menghasilkan y=7

Jadi, bagaimana Anda melatih jaringan saraf (neural network) untuk melakukan tugas yang sama? Menggunakan data! Dengan memberinya  satu set data Xs, dan satu set data Ys, sehingga kita bisa mengetahui hubungan antara keduanya.

Ini jelas merupakan paradigma yang sangat berbeda dari yang biasa Anda gunakan, jadi mari kita mulai


## Imports

Kita mulai dengan melakukan import library python. Kita melakukan import TensorFlow dan dipanggil dengan tf untuk memudahkan penggunaan.

kemudian kita mengimport library numpy, Untuk membantu kita merepresentasikan data dalam bentuk list lebih mudah dan cepat.

Framework untuk mendefinisikan neural network sebagai sequential layers dengan keras, jadi kita import juga.

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

print(tf.__version__)
```

## Define and Compile the Neural Network

Selanjutnya kita akan membuat neural network sederhana. Terdiri dari 1 layer, dan layer tersebut memiliki 1 neuron, dan input shape hanya 1 value.

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Sekarang kita akan compile neural network dengan menggunakan 2 functions, loss dan optimizer.

```python
model.compile(optimizer='sgd', loss='mean_squared_error')
```

## Providing the Data

Kemudian kita akan berikan data kedalam neural network. dalam kasus ini kita akan gunakan 6 data xs dan 6 data ys.

nilai x=-1, maka menghasilkan y=-3

nilai x=0, maka menghasilkan y=-1

nilai x=1, maka menghasilkan y=1

nilai x=2, maka menghasilkan y=3

nilai x=3, maka menghasilkan y=5

nilai x=4, maka menghasilkan y=7

Kita akan dibantu dengan numpy untuk mempermudah mendefinisikan list dalam python

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

# Training the Neural Network

Proses pelatihan jaringan saraf, di mana ia 'mempelajari' hubungan antara Xs dan Ys adalah dengan memanggil **model.fit** .  Di sinilah ia akan melalui perulangan sesuai dengan banyaknya epochs yang kita tentukan, membuat tebakan, mengukur seberapa baik atau buruknya (loss), menggunakan opimizer untuk membuat tebakan lain, dll. Ini akan melakukannya sebanyak epochs yang kita tentukan. Ketika Anda menjalankan kode ini, Anda akan melihat loss di sisi kanan.

```python
model.fit(xs, ys, epochs=1000)
```

Ok, sekarang kalian sudah memiliki model terlatih untuk mencari hubungan antara X and Y. Anda dapat menggunakan **model.predict** metode untuk mengetahui nilai Y dari nilai X. Jadi, sebagai contoh jika X = 10, Apakah nilai Y yang akan diberikan? kita akan lihat prediksi dari model kita:

```python
print(model.predict([10.0]))
```

Anda mungkin berpikir 19, kan? Tapi akhirnya menjadi sedikit di bawah. Menurut Anda mengapa demikian?

Ingat bahwa jaringan saraf berurusan dengan probabilitas, sehingga mengingat data yang kita berikan pada Neural Network, tetapi dengan hanya 6 data yang kita berikan. Akibatnya, hasil untuk x=10 sangat dekat dengan 19, tetapi belum tentu 19.

Saat Anda bekerja dengan jaringan saraf, Anda akan melihat pola ini berulang. Anda hampir selalu akan berurusan dengan probabilitas, bukan kepastian, dan akan melakukan sedikit pengkodean untuk mencari tahu apa hasilnya berdasarkan pada probabilitas, terutama ketika melakukan klasifikasi.
